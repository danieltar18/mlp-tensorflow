Prediction on test.csv


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
import tensorflow.keras.backend as K
import itertools
import json

from sklearn.model_selection import KFold
from sklearn.metrics import root_mean_squared_error, mean_absolute_error
```

```python
df = pd.read_csv("datasets/test.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.columns = [col.lower() for col in df]
```

```python
ids = df[["id"]]
df.drop("id", axis=1, inplace=True)
```

```python
df = df[df.select_dtypes(include="object").columns.to_list() + df.select_dtypes(exclude="object").columns.to_list()]
```

```python
X = tf.convert_to_tensor(df, dtype=tf.float32)
```

```python
with open("norm_layer_values", 'r') as f:
    params = json.load(f)
```

```python
norm_layer = tf.keras.layers.Normalization(axis=1, mean=params["mean"][0],  variance=params["variance"][0])
```

```python
X = norm_layer(X)
```

```python
mlp_model = tf.keras.models.load_model("mlp_model")
```

```python
val_predictions = mlp_model.predict(X)
```

```python
ids["medv"] = val_predictions
```

```python
ids.to_csv("submission.csv", index=False)
```
